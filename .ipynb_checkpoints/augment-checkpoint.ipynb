```python
import matplotlib.pyplot as plt
import numpy as np
import os
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
def plotimages(images_arr):
    fig,axes=plt.subplots(1,10,figsize=(20,20))
    axes=axes.flatten()
    for img ,ax in zip(images_arr,axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
gen=ImageDataGenerator(rotation_range=10,width_shift_range=0.1,height_shift_range=0.1,shear_range=0.15,zoom_range=0.1,channel_shift_range=10.,horizontal_flip=True,vertical_flip=True)
```

```python
chosen_image=random.choice(os.listdir('datasets/train/arduino_uno'))
```

```python
image_path='datasets/train/arduino_uno/'+chosen_image
```

```python
image=np.expand_dims(plt.imread(image_path),0)
plt.imshow(image[0])
```

```python
aug_iter=gen.flow(image,save_to_dir='datasets/augmented', save_prefix='aug-image-', save_format='jpeg')
```

```python
aug_images=[next(aug_iter)[0].astype(np.uint8) for i in range(10)]
```

```python
plotimages(aug_images)
```
